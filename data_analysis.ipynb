```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('transcriptions.csv')
```

```python
# Display the first few rows
print(df.head())

# Display summary statistics and info
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Plot histogram of transcription lengths
plt.figure(figsize=(10, 6))
plt.hist(df['transcription_length'], bins=20, edgecolor='black')
plt.title('Distribution of Transcription Lengths')
plt.xlabel('Transcription Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Transcription length statistics
df['transcription_length'] = df['transcription'].apply(len)
print(df['transcription_length'].describe())
```
